```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import os.path
import requests 
from os import path
```

<h3 style="text-align:left"> fetch_CIA_pages is a function which load the page on a local directory here : <b>scraped_CIA_Factbook</b><h3>

```python


directory = "scraped_CIA_Factbook"

if not path.isdir(directory):
    os.mkdir(directory)

    
#Reading of the html of the web page
def fetch_CIA_pages():
    
    file_name = directory +"/" +directory +".html"
    url = 'https://www.cia.gov/library/publications/the-world-factbook/'
    page = requests.get(url, allow_redirects=True)
    if not page.status_code == 200:
        print("some error occured")
    else:
        open(file_name, 'wb').write(page.content)
    return page
    


fetch_CIA_pages() 



```

<h3 style="text-align:left"> <b> process_factbook_pages </b> that processes all of the saved factbook "from html to soup"<h3>

```python
def process_factbook_pages():
    file_name = directory +"/" +directory +".html"
    
    with open(file_name, encoding= 'utf-8') as fp:
        soup = BeautifulSoup(fp, 'html.parser')
    #print(soup.prettify())
    return soup
    
process_factbook_pages()
```

<h3 style="text-align:left"> Retrieve the country's name and code's name. <h3>

```python
fact_CIA = process_factbook_pages()

#We access the drop-down menu which has the "option" for next acces data of each country

country = []
code_of_country = []

for info in fact_CIA.find_all('option'):
    code_of_country.append(info.get('value')[5:7]) # access to the code
    country.append(info.text.strip()) #acces to name of the countries
    
country.pop(0) # Deleting 'World'
code_of_country.pop(0) # Deleting 'World'
print(len(code_of_country))

  

    
print('Name of country\n'+'-'*50)
for name in country[1:]:
    print(name,end =',')
    
    
print('\n\n\nCode of Country\n'+'-'*50)
for code in code_of_country[1:]:
    print(code,end =' , ')
```

```python
#we put data in a dictionary

def data_to_dict():
    percentage_of_internet_user = {}
    
    for i in range(1, len(country) - 1):

        country_internet = country[i]

        txt = text_data[country_internet]
        pos = txt.find('Internet users:')

        if pos!=-1:
            pos = pos + len('Internet users:')
            string = txt[pos:pos+600]

            start = re.search('percent of population:', string)

            end = re.search('%', string)
            if (start != None and end!=None):
                start = start.end()
                end = end.start()
                a = string[start:end]
                if a[-1].isdigit():
                    a = float(a)
                    print(f"Internet users % of {country_internet}: {a}")
                    percentage_of_internet_user[country_internet] = a
                else:
                    print("no internet users data")
            else:
                print("no internet users data")
        else:
            print("no internet users data")

    print("\n Finished finding all Internet users data")
    
    return percentage_of_internet_user

            
```

```python
internet_users = data_to_dict()
```
